Q2. Plot a bar chart between brand and number of mobile of that brand.

```python
plt.figure(figsize = (15,10))
df["Brand"].value_counts().plot.bar(color=plt.cm.Paired(np.arange(len(df))))
plt.xlim(-1,20.5)

plt.xlabel("num of phones" , fontsize = 25)
plt.ylabel("Brand" , fontsize = 25)
plt.title("Brands ---- count of Models" , fontsize = 25)
plt.show()
```

Q3. Plot a chart between Brand and their median price "

```python
plt.figure(figsize = (35,8))
df1 = df.groupby(["Brand"])["Price"].median().reset_index()
sns.set(font_scale=2)
sns.barplot(x = df["Brand"] , y = df["Price"])
plt.xlim(-1,20.5)
plt.show()
```

Q5. Plot a scatterr chart between 'Battery capacity (mAh)' and 'Price' and give the color on 'Number of SIMs'

```python
plt.figure(figsize = (15,10))
sns.scatterplot(x = df["Battery capacity (mAh)"] , y = df["Price"] , hue = df["Number of SIMs"],
               palette = "deep")
```

Q6. Draw the correaltion heat map

```python
plt.figure(figsize = (15,10))
sns.heatmap(df.corr() , annot = True , mask = np.triu(np.ones_like(df.corr())))

plt.title("Corelation Matrix" , fontsize = 20)
plt.show()
```

 Q7. Draw a distribution chart for price"

```python
sns.kdeplot(df["Price"] , fill = True , color = "blue")
plt.show()
```
