Q2. Clean the data and remove the noise values from the data.

```python
df = df.replace("@" , np.nan)
```

```python
df = df.replace({"tt88":"88" , "89()":89})

```

```python
df["reading score"] = df["math score"].astype("float64")
df["math score"] = df["math score"].astype("float64")
```

```python
# selecting numerical featureb
df_num = df[["math score" , "reading score", "writing score"]]
```

```python
# using IQR method
# deletr noise and outlier
for i in df_num.columns:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3-Q1
    df = df[(df[i]>=(Q1-1.5*IQR)) & (df[i]<=(Q3+1.5*IQR))]
df
```

Q11. Create a dataframe where the index will be group type and columns will be minimum,
maximum, and average Maths score.



```python
df1 = df.groupby(["group"])["math score"].max().reset_index()

df1.rename(columns = {'math score':'max_math'}, inplace = True)
```

```python
df2 = df.groupby(["group"])["math score"].min().reset_index()
df2.rename(columns = {'math score':'min_math'}, inplace = True)
```

```python
df3 = df.groupby(["group"])["math score"].mean().reset_index()
df3.rename(columns = {'math score':'avg_math'}, inplace = True)
```

```python
df4 = pd.merge(df1,df2,on="group")
pd.merge(df4,df3 , on ="group")
```
