```python
# removing ","
data["No Claim bonus discount"] = data["No Claim bonus discount"].str.replace(",","")
data["Total IDV"] = data["Total IDV"].str.replace(",","")
data["Premium"] = data["Premium"].str.replace(",","")

 
```

```python
# Q1) Because of some high fraud rate, the company has decided not to give insurance to cars with
# RTO as Pen with CNG kit installed and older than 2005 (inclusive 2005). How much premium loss they
# will face because of this decision?
# Total_premium = data["Premium"].sum()
# Total_premium
loss = data["Premium"].sum() - data[(data["RTO"]=="Pen") & (data["Model_Year"]<=2005) & data["CNGkit"] == 1]["Premium"].sum()
loss
```

```python
# Which car manufacture company has major percentage in overall policy count in 2019?.
df1 = data[data["Model_Year"] == 2019].groupby(["Company"])["CustID"].count().reset_index().sort_values(
    by = "CustID" , ascending = False)
df1["Percentage"] = round(df1["CustID"]/df1["CustID"].sum()*100)
df1["Percentage"] = df1["Percentage"].astype(str)+"%"
 
df1[df1["Percentage"] == df1["Percentage"].max()]
df1
 
```

```python
# What is the correlation of premium and No Claim bonus discount for Maruti hatchback?
df2 = data[(data["Company"] == "Maruti") & (data["Car Type"] == "HatchBack")]
df2[["Premium" , "No Claim bonus discount"]].corr()
 
```

```python
#  What is the median of premium for sedan cars for the latest year?
data[(data["Model_Year"] == data["Model_Year"].max()) & (data["Car Type"] == "Sedan")]["Premium"].median()
```

```python
# What is the change percentage of the premium from 2018 to 2019 model for Pen RTO where IDV is 
# more than 8 lakh?

df3 = data[(data["RTO"] == "Pen") & (data["Total IDV"]>800000)&(data["Model_Year"]>2017)].groupby(["Model_Year"])["Premium"].sum().reset_index()
percent_change =str(round(((df3["Premium"][1]-df3["Premium"][0])/df3["Premium"][0])*100))+"%"
percent_change
 

 
```
