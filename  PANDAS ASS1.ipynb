```python
# drop constant value columns
for i in data.columns:
    if data[i].nunique()<=1:
        data.drop(columns = i , axis = 1 , inplace = True)
        
for i in data.columns[data.nunique()<=1]:
    data.drop(columns = i , axis = 1 , inplace = True)
 
     
```

```python
data["gender"].unique()
# only one record of other lets drop it
data.drop(data[data["gender"] == "Other"].index , inplace = True)
```

```python
sns.barplot(data = data.groupby(["gender"])["avg_glucose_level"].mean().reset_index(),
            x = "gender" , y = "avg_glucose_level")
plt.style.use("seaborn-notebook")
 
plt.title("Avg Suger level of gender")
plt.show()
```
