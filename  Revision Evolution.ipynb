```python
df_2015 = df[df["Year"]==2015]
df_2015.head()
```

Plot the region-wise average happiness index and dystopia residual score for 2015 and 2016

```python
plt.figure(figsize = (25,10))
sns.barplot(x = df_2015["Region"] ,y = df_2015["Happiness Score"])
plt.title("YEAR--2015" , fontsize = 50)
plt.tight_layout()
plt.show()
plt.figure(figsize = (25,10))
sns.barplot(x = df_2015["Region"] ,y = df_2015["Dystopia Residual"])
plt.show()
```

```python
plt.figure(figsize = (25,10))
sns.barplot(x = df_2016["Region"] ,y = df_2016["Happiness Score"])
plt.title("YEAR--2016" , fontsize = 50)
plt.tight_layout()
plt.show()
plt.figure(figsize = (25,10))
sns.barplot(x = df_2016["Region"] ,y = df_2016["Dystopia Residual"])
plt.show()
```

What are the top 5 countries with highest happiness index in 2015 and 2016 

```python
plt.figure(figsize = (10,5))
X = df_2015[["Country","Happiness Score"]].sort_values(by = "Happiness Score" , ascending = False)
X.plot(x = 'Country' , y =  "Happiness Score" , kind = "bar")
plt.title("YEAR--2015" , fontsize = 50)
plt.tight_layout()
plt.xlim(-1,4.5)
plt.show()
```

```python
plt.figure(figsize = (10,5))
X = df_2016[["Country","Happiness Score"]].sort_values(by = "Happiness Score" , ascending = False)
X.plot(x = 'Country' , y =  "Happiness Score" , kind = "bar")
plt.title("YEAR--2016" , fontsize = 50)
plt.tight_layout()
plt.xlim(-1,4.5)
plt.show()
```

What are the top 5 regions with lowest happiness index in 2015 and 2016 ?

```python
plt.figure(figsize = (10,5))
X = df_2015[["Country","Happiness Score"]].sort_values(by = "Happiness Score")
X.plot(x = 'Country' , y =  "Happiness Score" , kind = "bar")
plt.title("YEAR--2015" , fontsize = 50)
plt.tight_layout()
plt.xlim(-1,4.5)
plt.show()
```

```python
plt.figure(figsize = (10,5))
X = df_2016[["Country","Happiness Score"]].sort_values(by = "Happiness Score")
X.plot(x = 'Country' , y =  "Happiness Score" , kind = "bar")
plt.title("YEAR--2016" , fontsize = 50)
plt.tight_layout()
plt.xlim(-1,4.5)
plt.show()
```
